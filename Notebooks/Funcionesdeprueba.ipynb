```python
def cantidad_filmaciones_mes(df, mes: str):
    # Filtrar por el estado 'Released'
    df_released = df[df['status'] == 'Released']
    
    # Contar las películas por mes
    mes = mes.lower()
    cantidad = df_released[df_released['mes'] == mes].shape[0]
    return f"{cantidad} películas fueron estrenadas en el mes de {mes.capitalize()}"

# Probar la función con un mes de ejemplo
mes_a_probar = "abril"  # Cambia este valor al mes que quieras probar
resultado = cantidad_filmaciones_mes(df_combines, mes_a_probar)
print(resultado)
```

```python
def score_titulo(df, titulo_de_la_filmacion: str):
    # Filtrar por el título de la filmación
    film = df[df['title'].str.lower() == titulo_de_la_filmacion.lower()]
    
    if not film.empty:
        film = film.iloc[0]
        titulo = film['title']
        anio = film['release_date'].year
        score = film['popularity']
        return f"La película {titulo} fue estrenada en el año {anio} con un score/popularidad de {score}"
    else:
        return "Película no encontrada"

# Probar la función con un título de ejemplo
titulo_a_probar = "Toy Story"  # Cambia este valor al título que quieras probar
resultado = score_titulo(df_combines, titulo_a_probar)
print(resultado)
```

```python
def votos_titulo(df, titulo_de_la_filmacion: str):
    # Filtrar por el título de la filmación
    film = df[df['title'].str.contains(titulo_de_la_filmacion, case=False, na=False)]
    
    if not film.empty:
        film = film.iloc[0]
        if film['vote_count'] >= 2000:
            titulo = film['title']
            anio = film['release_date'].year
            total_votos = film['vote_count']
            promedio_votos = film['vote_average']
            return f"La película {titulo} fue estrenada en el año {anio}. La misma cuenta con un total de {total_votos} valoraciones, con un promedio de {promedio_votos}"
        else:
            return "La película no cumple con la condición de tener al menos 2000 valoraciones"
    else:
        return "Película no encontrada"

# Probar la función con un título de ejemplo
titulo_a_probar = "Toy Story"  # Cambia este valor al título que quieras probar
resultado = votos_titulo(df_combines, titulo_a_probar)
print(resultado)
```

```python
def get_actor(df, nombre_actor: str):
    # Filtrar por las filas que contienen información del elenco (cast)
    df_cast = df.dropna(subset=['cast'])

    # Crear variables para almacenar la cantidad de películas y el retorno total
    total_peliculas = 0
    total_retorno = 0

    # Iterar sobre las filas del dataframe
    for index, row in df_cast.iterrows():
        # Separar la cadena de actores por comas y verificar si el actor está en la lista
        cast_list = [actor.strip() for actor in row['cast'].split(',')]
        if nombre_actor.lower() in [actor.lower() for actor in cast_list]:
            total_peliculas += 1
            total_retorno += row['revenue'] - row['budget']

    # Calcular el promedio de retorno
    if total_peliculas > 0:
        promedio_retorno = total_retorno / total_peliculas
        return f"El actor {nombre_actor} ha participado de {total_peliculas} cantidad de filmaciones, el mismo ha conseguido un retorno de {total_retorno} con un promedio de {promedio_retorno} por filmación"
    else:
        return "Actor no encontrado o no tiene suficientes participaciones"

# Probar la función con un nombre de actor de ejemplo
actor_a_probar = "Tom Hanks"  # Cambia este valor al nombre del actor que quieras probar
resultado = get_actor(df_combines, actor_a_probar)
print(resultado)
```

```python
def get_director(df, nombre_director: str):
    # Filtrar por las filas que contienen información del equipo (crew)
    df_crew = df.dropna(subset=['crew'])

    # Crear variables para almacenar la cantidad de películas y el retorno total
    total_peliculas = 0
    total_retorno = 0
    peliculas_info = []

    # Iterar sobre las filas del dataframe
    for index, row in df_crew.iterrows():
        # Convertir la columna 'crew' a una lista de diccionarios
        try:
            crew_list = json.loads(row['crew'].replace("'", '"'))
        except json.JSONDecodeError:
            continue

        # Verificar si el director está en el equipo
        for crew_member in crew_list:
            if crew_member.get('department', '').lower() == 'directing' and crew_member.get('name', '').lower() == nombre_director.lower():
                total_peliculas += 1
                retorno_individual = row['revenue'] - row['budget']
                total_retorno += retorno_individual
                peliculas_info.append({
                    "titulo": row['title'],
                    "fecha_lanzamiento": row['release_date'],
                    "retorno_individual": retorno_individual,
                    "costo": row['budget'],
                    "ganancia": row['revenue']
                })

    # Calcular el promedio de retorno
    if total_peliculas > 0:
        promedio_retorno = total_retorno / total_peliculas
        return {
            "message": f"El director {nombre_director} ha participado de {total_peliculas} cantidad de filmaciones, el mismo ha conseguido un retorno de {total_retorno} con un promedio de {promedio_retorno} por filmación",
            "peliculas": peliculas_info
        }
    else:
        return "Director no encontrado o no tiene suficientes participaciones"

# Probar la función con un nombre de director de ejemplo
director_a_probar = "Steven Spielberg"  # Cambia este valor al nombre del director que quieras probar
resultado = get_director(df_combines, director_a_probar)
print(resultado)
```
