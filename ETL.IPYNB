```python
# FunciÃ³n para convertir strings en diccionarios si es necesario
def convert_to_dict(x):
    if pd.isnull(x):
        return {}
    try:
        return ast.literal_eval(x)
    except (ValueError, SyntaxError):
        return {}
```

```python
import ast
# Convertir strings en diccionarios
df['belongs_to_collection'] = df['belongs_to_collection'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else {})
```

```python

# Aplicar la funciÃ³n a la columna belongs_to_collection
df['belongs_to_collection'] = df['belongs_to_collection'].apply(convert_to_dict)
```

```python
# Crear nuevas columnas
df['collection_id'] = df['belongs_to_collection'].apply(lambda x: x.get('id') if isinstance(x, dict) else None)
df['collection_name'] = df['belongs_to_collection'].apply(lambda x: x.get('name') if isinstance(x, dict) else None)
df['collection_poster_path'] = df['belongs_to_collection'].apply(lambda x: x.get('poster_path') if isinstance(x, dict) else None)
df['collection_backdrop_path'] = df['belongs_to_collection'].apply(lambda x: x.get('backdrop_path') if isinstance(x, dict) else None)
```

```python
# FunciÃ³n para convertir strings en listas de diccionarios
def convert_to_list_of_dicts(x):
    if pd.isnull(x):
        return []
    try:
        return ast.literal_eval(x)
    except (ValueError, SyntaxError):
        return []

# Aplicar la funciÃ³n a la columna production_companies
df['production_companies'] = df['production_companies'].apply(convert_to_list_of_dicts)
```

```python
# FunciÃ³n para extraer informaciÃ³n de la lista de diccionarios
def get_company_info(companies, index, key):
    if isinstance(companies, list) and len(companies) > index:
        return companies[index].get(key, None)
    
    return None

# Crear nuevas columnas
df['production_company_1_id'] = df['production_companies'].apply(lambda x: get_company_info(x, 0, 'id'))
df['production_company_1_name'] = df['production_companies'].apply(lambda x: get_company_info(x, 0, 'name'))
df['production_company_2_id'] = df['production_companies'].apply(lambda x: get_company_info(x, 1, 'id'))
df['production_company_2_name'] = df['production_companies'].apply(lambda x: get_company_info(x, 1, 'name'))
```

```python
# Eliminar la columna original
df.drop(columns=['production_companies'], inplace=True)
```

```python
# FunciÃ³n para desanidar los gÃ©neros
def extract_genres(genre_list):
    if pd.isna(genre_list) or genre_list == '[]':
        return ''
    else:
        # Reemplazar comillas simples por comillas dobles
        genre_list = genre_list.replace("'", '"')
        genres = json.loads(genre_list)
        return ', '.join([genre['name'] for genre in genres])

# Aplicar la funciÃ³n a la columna 'genres'
df['genres'] = df['genres'].apply(extract_genres)

# Mostrar el DataFrame resultante
print(df)
```

```python
# FunciÃ³n para desanidar y manejar errores
def extract_cast_or_crew(column):
    if pd.isna(column) or column == '[]':
        return np.nan
    else:
        try:
            items = json.loads(column.replace("'", '"'))
            return ', '.join([item['name'] for item in items])
        except json.JSONDecodeError:
            return np.nan  # Reemplazar con NaN si hay un error en el JSON
```

```python
# Aplicar la funciÃ³n a las columnas 'cast' y 'crew'
df_credits['cast'] = df_credits['cast'].apply(extract_cast_or_crew)
df_credits['crew'] = df_credits['crew'].apply(extract_cast_or_crew)
```

```python
# Contar los valores NaN
cast_errors = df_credits['cast'].isna().sum()
crew_errors = df_credits['crew'].isna().sum()
```

```python
# Mostrar los valores NaN
print(f"Errores de JSON en 'cast' (NaN): {cast_errors}")
print(f"Errores de JSON en 'crew' (NaN): {crew_errors}")
```
