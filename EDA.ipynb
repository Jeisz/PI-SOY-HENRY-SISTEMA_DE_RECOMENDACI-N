```python
# FunciÃ³n para desanidar y manejar errores
def extract_cast_or_crew(column):
    if pd.isna(column) or column == '[]':
        return []
    else:
        try:
            items = json.loads(column.replace("'", '"'))
            return [item['name'] for item in items]
        except json.JSONDecodeError:
            return []
```

```python
def explorar_columna_crew(df, num_filas=5):
    # Imprimir algunas filas de la columna 'crew'
    print(df['crew'].head(num_filas))

# Ejecutar la funciÃ³n para explorar la columna 'crew'
explorar_columna_crew(df_credits, num_filas=10)
```

```python
def separar_columna_crew(df):
    # Crear una lista para almacenar los datos expandidos
    crew_data = []

    # Iterar sobre las filas del dataframe
    for index, row in df.iterrows():
        # Convertir la columna 'crew' a una lista de diccionarios
        try:
            crew_list = json.loads(row['crew'].replace("'", '"'))
        except json.JSONDecodeError:
            continue
        
        # Agregar los datos expandidos a la lista
        for crew_member in crew_list:
            crew_member['movie_id'] = row['id']  # Agregar el ID de la pelÃ­cula para referencia
            crew_data.append(crew_member)

    # Crear un nuevo dataframe a partir de la lista de datos expandidos
    df_crew_expanded = pd.DataFrame(crew_data)
    return df_crew_expanded

# Separar la columna 'crew' en el dataframe 'df_credits'
df_crew_expanded = separar_columna_crew(df_credits)

# Mostrar las primeras filas del nuevo dataframe
df_crew_expanded.head()
```

```python
# Separar la columna 'crew' en el dataframe 'df_credits'
df_crew_expanded = separar_columna_crew(df_credits)

# Renombrar la columna 'name' a 'director'
df_crew_expanded.rename(columns={'name': 'director'}, inplace=True)

# Mostrar las primeras filas del nuevo dataframe
df_crew_expanded.head()
```

```python
# Realizar la uniÃ³n (merge) de los dataframes en base a las columnas 'movie_id' y 'id'
df_merged = pd.merge(df_combines, df_crew_expanded, left_on='id', right_on='movie_id', how='left')

# Mostrar las primeras filas del dataframe resultante
df_merged.head()
```

```python
# VisualizaciÃ³n de datos faltantes en df_combined
plt.figure(figsize=(10, 6))
sns.heatmap(df_merged.isnull(), cbar=False, cmap='viridis')
plt.title('Datos faltantes en df_combined')
plt.show()
```

```python
# Verificar la cantidad de datos faltantes por columna
df_merged.isnull().sum()
```

```python
# Eliminar columnas con mÃ¡s del 50% de datos faltantes
threshold = len(df_merged) * 0.5
df_cleaned = df_merged.dropna(thresh=threshold, axis=1)
```

```python
# Eliminar filas con una cantidad significativa de datos faltantes
df_cleaned = df_cleaned.dropna(thresh=len(df_cleaned.columns) * 0.5, axis=0)
```

```python
# Imputar datos faltantes con la mediana (para columnas numÃ©ricas)
num_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
df_cleaned[num_cols] = df_cleaned[num_cols].fillna(df_cleaned[num_cols].median())
```

```python
# Imputar datos faltantes con la moda (para columnas categÃ³ricas)
cat_cols = df_cleaned.select_dtypes(include=['object']).columns
df_cleaned[cat_cols] = df_cleaned[cat_cols].fillna(df_cleaned[cat_cols].mode().iloc[0])
```

```python
def get_director(df, nombre_director: str):
    # Filtrar por las filas que contienen informaciÃ³n del equipo (crew)
    df_director = df[df['director'].str.lower() == nombre_director.lower()]

    # Crear variables para almacenar la cantidad de pelÃ­culas y el retorno total
    total_peliculas = df_director.shape[0]
    total_retorno = (df_director['revenue'] - df_director['budget']).sum()
    peliculas_info = df_director[['title', 'release_date', 'revenue', 'budget']].copy()
    peliculas_info['retorno_individual'] = peliculas_info['revenue'] - peliculas_info['budget']
    peliculas_info.rename(columns={'title': 'titulo', 'release_date': 'fecha_lanzamiento', 'revenue': 'ganancia', 'budget': 'costo'}, inplace=True)

    # Calcular el promedio de retorno
    if total_peliculas > 0:
        promedio_retorno = total_retorno / total_peliculas
        return {
            "message": f"El director {nombre_director} ha participado de {total_peliculas} cantidad de filmaciones, el mismo ha conseguido un retorno de {total_retorno} con un promedio de {promedio_retorno} por filmaciÃ³n",
            "peliculas": peliculas_info.to_dict(orient='records')
        }
    else:
        return "Director no encontrado o no tiene suficientes participaciones"

# Probar la funciÃ³n con un nombre de director de ejemplo
director_a_probar = "Steven Spielberg"  # Cambia este valor al nombre del director que quieras probar
resultado = get_director(df_merged, director_a_probar)
print(resultado)
```
